## Transformada de Fourier
La Transformada de Fourier es una herramienta matematica que permite analizar una señal en el dominio de la frecuencia. Su función es descomponer una señal compleja en una sumatoria de ondas sinusoidales más simples. 

Expresión matemática para señales continuas: 

$X(f)=∫−∞∞​x(t)e−j2πftdt$ 

Donde:

- x(t) es la señal original en el tiempo.
- X(f) es la representación de la señal en frecuencia.
- f corresponde a la frecuencia.
- $e^-j2πft$ representa las ondas sinusoidales utilizadas para analizar la señal.
  
Transformada Discreta de Fourier (DFT)

Los audios digitales están formados por un conjunto finito de muestras tomadas a intervalos regulares. Debido a que la Transformada de Fourier continua trabaja con señales continuas en el tiempo, no puede aplicarse directamente a señales digitales.

Por esta razón se utiliza la Transformada Discreta de Fourier (DFT), que permite analizar el contenido frecuencial de una señal muestreada. 

Expresión matemática en señales digitales: 

$X[k]=∑n=0N−1​x[n]e−jN2π​kn$

Donde:

- x[n]: muestra número n de la señal.
- N: número total de muestras.
- k: índice de frecuencia analizada.
- X[k]: valor complejo asociado a la frecuencia k.

Transformada Discreta de Fourier Inversa (IDFT)

Después de analizar una señal mediante la DFT y obtener sus componentes de frecuencia, es posible regresar al dominio temporal utilizando la Transformada Discreta de Fourier Inversa (IDFT).

La IDFT permite reconstruir la señal original a partir de la información de frecuencia obtenida previamente.

Expresión matemática de la inversa: 

$x[n]=N1​∑k=0N−1​X[k]ejN2π​kn$

Donde:

- x[n]: muestra reconstruida de la señal en el tiempo.
- X[k]: componentes de frecuencia obtenidas mediante la DFT.
- N: número total de muestras.
- k: índice de frecuencia.
- n: índice temporal.
